# Data Modelling

### Normalize tables in 3NF to apply STAR Schema.

### Dimensional Product

```python
dim_product = df_sales[['ProductKey', 'Product Name', 'Brand', 'Color', 'Subcategory', 'Category']].drop_duplicates()
```

### Dimensional Customer

```python
dim_customer = df_sales[['CustomerKey', 'Customer Code', 'Name', 'Education', 'Occupation']].drop_duplicates()
```

### Dimensional Region

```python
dim_region = df_sales[['Continent', 'City', 'State', 'CountryRegion']].drop_duplicates().reset_index(drop=True)
dim_region.insert(0, 'RegionKey', range(1, len(dim_region) + 1))
```

### Dimensional Date

```python
dim_date = df_sales[['OrderDate']].drop_duplicates().reset_index(drop=True)
dim_date.insert(0, 'DateKey', range(1, len(dim_date) + 1))

dim_date['Year'] = dim_date['OrderDate'].dt.year
dim_date['Month'] = dim_date['OrderDate'].dt.month
dim_date['MonthName'] = dim_date['OrderDate'].dt.month_name()
dim_date['Day'] = dim_date['OrderDate'].dt.day
dim_date['DayName'] = dim_date['OrderDate'].dt.day_name()

unique_years = sorted(dim_date['Year'].unique())
year_mapping = {year: i + 1 for i, year in enumerate(unique_years)}
dim_date['year_id'] = dim_date['Year'].map(year_mapping).astype(int)
```

### Fact Tables 

```python
df_sales_mapped = df_sales.merge(dim_region, on=['Continent', 'City', 'State', 'CountryRegion'], how='left')
df_sales_mapped = df_sales_mapped.merge(dim_date, on='OrderDate', how='left')

fact_sales = df_sales_mapped[['DateKey', 'ProductKey', 'CustomerKey', 'RegionKey', 'Quantity', 'Net Price']]
fact_sales['SalesAmount'] = fact_sales['Quantity'] * fact_sales['Net Price']


fact_forecast = df_forecast[['CountryRegion', 'Brand', 'Forecast', 'Year']].drop_duplicates().reset_index(drop=True)
fact_forecast['year_id'] = fact_forecast['Year'].map(year_mapping).astype(int)
```

### Saving Tables to CSV

```python
dim_region.to_csv('data/dim_region.csv', index=False)
dim_product.to_csv('data/dim_product.csv', index=False)
dim_customer.to_csv('data/dim_customer.csv', index=False)
dim_date.to_csv('data/dim_date.csv', index=False)

fact_sales.to_csv('data/fact_sales.csv', index=False)
fact_forecast.to_csv('data/fact_forecast.csv', index=False)
```
